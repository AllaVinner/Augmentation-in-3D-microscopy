```python
import skimage.io
import jview
import os
from collections import defaultdict
import matplotlib.pyplot as plt
```

```python
dataset = 'SAM'
DATASET_PATH = os.path.join('/scratch/ottosson/datasets', dataset)
LABEL_PATH = os.path.join(DATASET_PATH, 'label')
```

```python
print("Label path exist: ", os.path.isdir(LABEL_PATH))
print(f"Label dir has {len(os.listdir(LABEL_PATH))} files")
```

# What is in the dir

```python
files = os.listdir(LABEL_PATH)
num_file_types = defaultdict(int)
for f in files:
    num_file_types[jview.get_extention(f)] +=1
print("Number of different files (based on extentions)")
jview.print_dir(num_file_types)
```

# What is in the files?

```python
ndims = []
shapes = []
dtypes = []
mins = []
maxs = []
for f in files:
    if jview.get_extention(f) != 'tif': continue
    im = skimage.io.imread(os.path.join(LABEL_PATH, f))
    ndims.append(im.ndim)
    shapes.append(im.shape)
    dtypes.append(im.dtype)
    mins.append(im.min())
    maxs.append(im.max())
```

```python
print("Ndims: ", ndims)
print("dtypes: ", dtypes)
print("mins: ", mins)
print("maxs: ", maxs)
print("shapes: \n", shapes)
```
