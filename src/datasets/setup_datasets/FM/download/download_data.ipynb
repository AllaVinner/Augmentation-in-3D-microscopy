# Download data
This notebook downloads the data from the internet with 'wget' and stores it in the folder 'FM'.

```python
#imports
import os
import zipfile 
```

```python
download_link = r"https://www.repository.cam.ac.uk/bitstream/handle/1810/318119/wt.zip?sequence=8&isAllowed=y"
os.system('wget -O datazip.zip ' + download_link)
```

```python
with zipfile.ZipFile('datazip.zip', 'r') as zip_ref:
    zip_ref.extractall('data')
os.system('rm datazip.zip')
```
